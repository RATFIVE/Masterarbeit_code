```python
import pandas as pd
import numpy as np



v1_dl = "../thesis_plots/Versuch1_DL_LSTM_results.csv"
v2_dl = "../thesis_plots/Versuch2_DL_LSTM_results.csv"
v3_dl = "../thesis_plots/Versuch3_DL_results_classification=False_lightmode=False.csv"
v1_ml = "../thesis_plots/Versuch1_ML_model_statistics_all_models.csv"
v2_ml = "../thesis_plots/Versuch2_ML_model_statistics_all_models.csv"
v3_ml = "../thesis_plots/Versuch3_ML_model_statistics_all_models.csv"


```

# Versuch1

```python
def custom_score(alpha, recall, MSE):
    return alpha * (1 - recall) + (1 - alpha) * MSE

df_dl = pd.read_csv(v1_dl).rename(columns={
    'Num Parameters': 'Num Params',
    'Time Prediction [s]': 'Avg. Predict Time (s)'
})
df_dl['Model'] = 'LSTM'
df_dl['Custom Score'] = df_dl.apply(lambda row: custom_score(0.7, row['Recall'], row['MSE']), axis=1)
display(df_dl)

df_ml = pd.read_csv(v1_ml)
df_ml['Custom Score'] = df_ml.apply(lambda row: custom_score(0.7, row['Recall'], row['MSE']), axis=1)
display(df_ml)

df_results = pd.concat([df_dl, df_ml], axis=0).sort_values(by='RMSE', ascending=True)
df_results = df_results.reset_index(drop=True)
columns_to_round = ['RMSE', 'MSE', 'MAE', 'Recall', 'Custom Score', 'Accuracy', 'Precision', 'Avg. Fit Time (s)', 'Avg. Predict Time (s)', 'Num Params']
df_results[columns_to_round] = df_results[columns_to_round].round(4)

display(df_results[['Model', 'RMSE', 'MSE', 'MAE', 'Recall', 'Custom Score', 'Accuracy', 'Precision','Avg. Fit Time (s)','Avg. Predict Time (s)', 'Num Params']].reset_index(drop=True))

mean_rmse_v1 = df_results['RMSE'].mean()
mean_score_v1 = df_results['Custom Score'].mean()
```
