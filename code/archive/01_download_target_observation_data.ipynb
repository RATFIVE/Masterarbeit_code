```python
import xarray as xr
import numpy as np
import pandas as pd
import plotly.graph_objects as go

```

```python
# Open 

f = 'data/NO_TS_TG_KielHoltenauTG.nc'
ds = xr.open_dataset(f)
df = ds.to_dataframe().reset_index()
```

```python
df.info()
df.head()
```

```python
# Just take quality check 1 data
print(df['SLEV_QC'].unique())

df_qc = df[df['SLEV_QC'] == 1]
df_qc.info()
```

```python
# Group the Data hurly and take the mean
df_hourly = df_qc.copy()
df_hourly.set_index('TIME', inplace=True)
display(df_hourly.head())

df_hourly = df_hourly.resample('h').mean(numeric_only=True)
df_hourly.reset_index(inplace=True)
df_hourly.info()
df_hourly.head()
```

```python
# Just take the data from 2023-01-01 
df_hourly = df_hourly[df_hourly['TIME'] > '2023-01-01']
df_hourly.sort_values('TIME', inplace=True)
df_hourly.reset_index(inplace=True, drop=True)
df_hourly.info()
df_hourly.head()
```

```python
# Save Target Data
station = df['STATION'].unique()[0]
filename = f'{station}_hourly'

df_hourly.to_json(f'data/{filename}.json', orient='records', lines=True, date_format='iso')
```
