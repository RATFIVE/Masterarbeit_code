```python
import os
import datetime
import pandas as pd
import xarray as xr
from tqdm import tqdm
from joblib import Parallel, delayed

files = [os.path.join(file_path, file) for file in os.listdir(file_path) if '_FB_' not in file]

def process_file(f):
    try:
        ds = xr.open_dataset(f)
        df = ds.to_dataframe().reset_index()

        # Filtere NaN-Spalten/-Zeilen
        df.dropna(axis=1, how='all', inplace=True)
        df.dropna(axis=0, how='all', inplace=True)

        # Zeitfilter
        if df['TIME'].min() < datetime.datetime(2014, 1, 1) or df['TIME'].max() < datetime.datetime(2025, 2, 1):
            return None  # Datei ignorieren
        
        # Filter Data if it is outside the BBOX
        df = df[
            (BBOX['minimum_latitude'] <= df['LATITUDE']) & (df['LATITUDE'] <= BBOX['maximum_latitude']) &
            (BBOX['minimum_longitude'] <= df['LONGITUDE']) & (df['LONGITUDE'] <= BBOX['maximum_longitude'])
        ]
        if df.empty:
            return None

        # Filter Data from datetime.datetime(1980, 1, 1) to datetime.datetime(2025, 3, 1)
        df = df[(df['TIME'] >= datetime.datetime(2014, 1, 1)) & (df['TIME'] <= datetime.datetime(2025, 2, 1))]
        if df.empty:
            return None  # Datei ignorieren

        # Station extrahieren
        try:
            station = df['STATION'].unique()[0]
        except Exception as e:
            print(f"Fehler beim Extrahieren der Station in {f}: {e}")
            return None

        # Zeit auf ganze Stunden runden
        df.set_index('TIME', inplace=True)
        df = df.resample('h').mean(numeric_only=True)
        df['STATION'] = station
        df.reset_index(inplace=True)

        return df
    except Exception as e:
        print(f"Fehler bei Datei {f}: {e}")
        return None

# Parallele Verarbeitung mit 8 Worker-Prozessen
num_workers = -1  # Anpassen je nach CPU-Kerne
df_list = Parallel(n_jobs=num_workers)(
    delayed(process_file)(file) for file in tqdm(files, desc="Processing Files")
)

# Nur nicht-leere DataFrames behalten
df_list = [df for df in df_list if df is not None]

# Zusammenführen der Ergebnisse
df_combined = pd.concat(df_list, ignore_index=True)

print(df_combined.info())

```

```python
from joblib import Parallel, delayed
import pandas as pd
from tqdm import tqdm

# Einzigartige Zeitstempel
df_merged = pd.DataFrame({'TIME': df_combined['TIME'].unique()})

def process_coordinate(coord):
    """ Verarbeitet eine einzelne Koordinate """
    ocean_sub_data = df_combined.loc[df_combined['POSITION'] == coord]
    
    if ocean_sub_data.empty:
        return None  # Falls keine Daten vorhanden sind

    station = ocean_sub_data['STATION'].unique()[0]
    station = station.decode('utf-8') if isinstance(station, bytes) else str(station)

    ocean_sub_data = ocean_sub_data.drop(columns=['LATITUDE', 'LONGITUDE'])
    
    cols = [col for col in ocean_sub_data.columns if col not in ['TIME', 'POSITION']]
    
    ocean_sub_data = ocean_sub_data.rename(columns={col: f"{col}-{station}-{coord}" for col in cols})
    
    ocean_sub_data = ocean_sub_data.drop(columns=['POSITION'])

    return ocean_sub_data

# Parallelisierung mit tqdm-Integration
results = Parallel(n_jobs=-1, backend="loky", timeout=1000)(
    delayed(process_coordinate)(coord) for coord in tqdm(coordinates, desc="Processing Coordinates")
)

# Entferne None-Werte
results = [df for df in results if df is not None]


```

```python
# Zusammenfügen der Ergebnisse
df_merged = df_merged.copy()
for ocean_sub_data in tqdm(results, desc="Merging Data"):
    df_merged = df_merged.merge(ocean_sub_data, on='TIME', how='left')

display(df_merged.info())
```

```python
df_all = pd.DataFrame()
end_date = pd.to_datetime("2025-03-01")
start_date = pd.to_datetime("2025-01-01")

for file in tqdm(files, desc='Processing files'):
    f = os.path.join("data", file)
    ds = xr.open_dataset(f)
    df = ds.to_dataframe().reset_index()
    df['TIME'] = pd.to_datetime(df['TIME'])
    # Filtern nach BBOX
    df = df[
        (df['LATITUDE'] >= BBOX['minimum_latitude']) &
        (df['LATITUDE'] <= BBOX['maximum_latitude']) &
        (df['LONGITUDE'] >= BBOX['minimum_longitude']) &
        (df['LONGITUDE'] <= BBOX['maximum_longitude'])
    ]

    # Filter das dataframe nach dem Startdatum
    df = df.loc[df['TIME'] >= start_date]

    # wenn das DataFrame daten bis zum 01.03.2025 enthält, dann füge es zum df_all hinzu
    if (df['TIME'].max() >= start_date) & (df['TIME'].max() <= end_date):
        df_all = pd.concat([df_all, df], axis=0)



    
print(df_all['STATION'].nunique())
```
